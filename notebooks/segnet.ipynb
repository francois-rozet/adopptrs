```python
import random
import via as VIA

from torch.utils.data import DataLoader
from dataset import VIADataset, ToTensor

via = VIA.load('../products/json/california.json')

keys = sorted(list(via.keys()))

random.seed(0)
random.shuffle(keys)

valid_via = {key: via[key] for i, key in enumerate(keys) if (i % 5) == 1} # fold 1
validset = ToTensor(VIADataset(valid_via, '../resources/california/', shuffle=True))
validloader = DataLoader(validset, batch_size=10)
```

# SegNet

`SegNet` was trained with the `DiceLoss` loss function for `20` epochs.

```python
import torch
from torchsummary import summary
from models import SegNet

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = SegNet(3, 1).to(device)
model.load_state_dict(torch.load('../products/models/segnet_1_020.pth', map_location=device))

summary(model, input_size=(3, 256, 256))
```

## Training

```python
import matplotlib.pyplot as plt
import pandas as pd

stats = pd.read_csv('../products/csv/segnet.csv')

plt.figure(figsize=(8, 6))

for i in range(5):
    plt.plot(
        stats[stats['model'] == 'segnet_{}'.format(i)]['epoch'],
        stats[stats['model'] == 'segnet_{}'.format(i)]['train_loss_mean'],
        label='{}'.format(i)
    )

plt.xlabel('Epoch')
plt.ylabel('Mean loss')
plt.legend(title='Fold')
plt.show()
```

```python
plt.figure(figsize=(8, 6))

for i in range(5):
    plt.plot(
        stats[stats['model'] == 'segnet_{}'.format(i)]['epoch'],
        stats[stats['model'] == 'segnet_{}'.format(i)]['train_loss_second'],
        label='{}'.format(i)
    )

plt.xlabel('Epoch')
plt.ylabel('Median loss')
plt.legend(title='Fold')
plt.show()
```

## Sample

```python
from dataset import to_pil
from plots import plot_alongside

model.eval()

with torch.no_grad():
    for inputs, targets in validloader:
        break

    outputs = model(inputs.to(device)).cpu()

plot_alongside(
    map(to_pil, inputs),
    map(to_pil, targets),
    map(to_pil, outputs),
    zoom=5
)

plt.tight_layout()
```
